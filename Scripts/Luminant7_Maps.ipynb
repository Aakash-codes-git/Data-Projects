```python
scope = ['New Jersey']


values = df_sample['Total Crashes'].tolist()
fips = df_sample['fip_codes'].tolist()

#colorscale = ["#8dd3c7", "#ffffb3", "#bebada", "#fb8072",
              #"#80b1d3", "#fdb462", "#b3de69", "#fccde5",
              #"#d9d9d9", "#bc80bd", "#ccebc5", "#ffed6f",
              #"#8dd3c7", "#ffffb3", "#bebada", "#fb8072",
              #"#80b1d3", "#fdb462", "#b3de69", "#fccde5",
              #"#d9d9d9"]
binning_endpoints = list(np.mgrid[min(values):max(values):11j])
colorscale = [
    'rgb(239,110,110)',
    'rgb(193,100,100)',
    'rgb(195,90,90)',
    'rgb(144,80,80)',
    'rgb(101,70,70)',
    'rgb(85,60,60)',
    'rgb(65,50,50)',
    'rgb(40,40,40)',
    'rgb(30,30,30)',
    'rgb(22,20,20)',
    'rgb(15,10,10)',
    'rgb(8,2,2)'
]

fig = ff.create_choropleth(
    fips=fips, values=values, scope=scope, binning_endpoints=binning_endpoints,
    colorscale=colorscale, round_legend_values=True,
    simplify_county=0, simplify_state=0,
    county_outline={'color': 'rgb(5, 10, 20)', 'width': 0.8},
    state_outline={'width': 1},
    legend_title='Total Crashes',
    title='2001 Crashes by NJ_Counties',height=500, width=900
)

fig.layout.template = None
fig.show()
```
